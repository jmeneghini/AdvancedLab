# Exercise 1
RA vs Dec of the 5000 brightest stars in the sky from PhotoTag database

```python
# Importing the data
ex1 = pd.read_csv('../data/exercise1Data.csv', skiprows=2, header = None, names = ['ra', 'dec', 'mag'])
```

```python
# Plotting the data
fig, ax = plt.subplots(figsize=plots.set_size('default', fraction=1))
# set markers to circles
ax.scatter(ex1['ra'], ex1['dec'], s=0.1)
ax.set_xlabel('RA (deg)')
ax.set_ylabel('Dec (deg)')
ax.set_title('RA vs Dec of the 5000 brightest stars in the sky')
plt.savefig('../figures/ex1.pdf')
```

# Exercise 2
Two color plot of stars with modelMag_g between 18 and 19, from the area of the sky with 0.0 < RA < 5.0 and 0.0 < Dec < 5.0. Colors should be (modelMag_u - modelMag_g) vs (modelMag_g - modelMag_r).

```python
# Importing the data
ex2 = pd.read_csv('../data/exercise2Data.csv', skiprows=2, header = None, names = ['u', 'g', 'r'])
```

```python
# Plotting the data
fig, ax = plt.subplots(figsize=plots.set_size('default', fraction=1))
ug = ex2['u'] - ex2['g']
gr = ex2['g'] - ex2['r']
ax.scatter(ug, gr, s=0.1)
ax.set_xlabel('u-g')
ax.set_ylabel('g-r')
ax.set_title("Two Color Plot of Stars")
ax.set_xlim(-1, 3.5)
ax.set_ylim(-1, 2)
plt.savefig('../figures/ex2.pdf')
```

# Exercise 4
Make a histogram of the available redshifts of the galaxies contained in the sky region with 179.0 < RA < 182.0 and 28.0 < Dec < 32.0

```python
# Importing the data
ex4 = pd.read_csv('../data/exercise4Data.csv', skiprows=2, header = None, names = ['z'])
ex4
```

```python
# Plotting the data
fig, ax = plt.subplots(figsize=plots.set_size('default', fraction=1))
# set bin width to 0.001
num_bins = int(0.2/0.001)
ax.hist(ex4['z'], bins = num_bins)
ax.set_xlabel('z')
ax.set_ylabel('Number of galaxies')
ax.set_title('Histogram of the available redshifts of the galaxies')
plt.savefig('../figures/ex4.pdf')
```

# Exercise 5
Make a pie-plot diagram of the RA-redshift distribution of all the galaxies with 1.0 degree of the celestial equator (Dec = 0.0)

```python
# Importing the data
ex5 = pd.read_csv('../data/exercise5Data.csv', skiprows=2, header = None, names = ['ra', 'z'])
```

```python
# Plotting the data
fig, ax = plt.subplots(subplot_kw={"projection": "polar"} ,figsize=plots.set_size('default', fraction=1))
ax.plot(ex5['ra']*np.pi/180, ex5['z'], 'o', markersize=0.1)
ax.set_title('RA-redshift distribution of all the galaxies')
plt.savefig('../figures/ex5.pdf')
```
